```python
def prior_times_country_data():
    import pandas as pd
    import matplotlib.pyplot as plt
    import numpy as np

    # this gives the country level aggregation by sector
    # the file used here 
    sector_country = pd.read_csv('~/Desktop/Founders_Pledge/affectable_emissions_active_repo/Affectable-Emissions/data_files/inputs_w_country_fix/tong_cntry_fix.csv', index_col = 0).iloc[:,:9]

    # this gives our prior for the weight of a given year in each sector (for CO2 emissions)
    # (this isn't really the same for every country, but we're going to approximate it like it is)
    p_dat = pd.read_csv('~/Desktop/Founders_Pledge/affectable_emissions_active_repo/Affectable-Emissions/data_files/tong_commited_emit_data_weighted/tong_prior.csv', usecols = ['Year','Electricity.1', 'Industry.1', 'Road Transport.1', 'Other Transport.1',
           'International Tranpsort.1', 'Residential.1', 'Commerical.1',
           'Other Energy.1'])

    #this cleans the prior dat, makes it easier to multiply
    p_dat.columns = sector_country.columns
    sector_country = sector_country.rename(columns = {'country.name.en':'country'})
    p_dat = p_dat.rename(columns = {'Unnamed: 0':'Year'})


    fs = []
    for x in range(len(sector_country)):
        country_prior_yr_sector = p_dat.iloc[:,1:].multiply(sector_country.iloc[:,1:].loc[x], axis = 1)
        country_prior_yr_sector.iloc[:,0] = p_dat.iloc[:,0]
        country_prior_yr_sector['country'] = sector_country['country'].loc[x]
        country_prior_yr_sector = country_prior_yr_sector.rename(columns = {'Electricity':'Year'})
        fs.append(country_prior_yr_sector)

    country_tong_commit_data = pd.concat(fs)
    country_tong_commit_data.to_csv('~/Desktop/Founders_Pledge/affectable_emissions_active_repo/Affectable-Emissions/data_files/tong_commited_emit_data_weighted/country_level_posteriors/all_countriesxprior.csv')

    all_country_weights = country_tong_commit_data.copy()
    all_country_weights.iloc[:,1:8] = 1.0
    #all_country_weights.iloc[10,1] = 500
    all_country_weights.to_csv('~/Desktop/Founders_Pledge/affectable_emissions_active_repo/Affectable-Emissions/data_files/tong_commited_emit_data_weighted/country_level_posteriors/all_countries_weights.csv')

prior_times_country_data()
```
