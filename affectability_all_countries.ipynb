```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
from collections import Counter
```

```python
all_emit_data = pd.read_csv('~/Desktop/Founders_Pledge/affectable_emissions_active_repo/Affectable-Emissions/data_files/master_file/affectable_emissions_MASTER.csv', index_col = 0)

# make a makeshift affectability file according to this format
affectability = all_emit_data.copy()
for col in affectability.columns[4:]:
    affectability[col].values[:] = 0.05
affectability.to_csv('~/Desktop/Founders_Pledge/affectable_emissions_active_repo/Affectable-Emissions/data_files/affectability_data/all_countries_placeholder.csv')
```

```python
# make sure the columns of the dfs are the same before multiplying by index
assert affectability.columns.equals(all_emit_data.columns) == True

# multiply the dfs columnwise, weighting emissions by their affectibility factor
affectable_emissions = affectability.iloc[:,4:].mul(all_emit_data.iloc[:,4:], axis = 1)
affectable_emissions[['country', 'region', 'SSP', 'RCP']] = affectability.iloc[:,:4]

# put country, rcp, ssp columns at the front of the df
country = affectable_emissions.pop('country')
ssp = affectable_emissions.pop('SSP')
rcp = affectable_emissions.pop('RCP')
reg = affectable_emissions.pop('region')
affectable_emissions.insert(0, country.name, country)
affectable_emissions.insert(1, reg.name, reg)
affectable_emissions.insert(2, ssp.name, ssp)
affectable_emissions.insert(3, rcp.name, rcp)

affectable_emissions.to_csv('~/Desktop/Founders_Pledge/affectable_emissions_active_repo/Affectable-Emissions/data_files/affectability_data/all_countries_emit_weighted_by_affectability.csv')
```
